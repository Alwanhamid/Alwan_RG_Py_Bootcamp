# Final Project Guideline

## Goals
Peserta diminta untuk membuat **aplikasi Python interaktif berbasis Streamlit** dengan **minimal satu fitur AI** atau **fungsionalitas interaktif** lainnya. Proyek ini bertujuan untuk menunjukkan penerapan semua materi yang telah dipelajari selama program.



## Kriteria Project
- Menggunakan **Python** sebagai bahasa utama.
- Menggunakan **Streamlit** sebagai UI aplikasi.
- Minimal **1 fitur interaktif** (input form, tombol, chatbot, dll).
- Wajib menggunakan minimal **1 library eksternal** selain `streamlit`.
- Disarankan menerapkan konsep-konsep Python yang telah dipelajari, seperti: OOP, Data Structures, Looping, etc.
- **Deploy** aplikasi ke platform seperti:
  - [Streamlit Cloud](https://share.streamlit.io/)
  - [HuggingFace Space](https://huggingface.co/spaces)
- Sertakan **dokumentasi proyek** (README.md atau PDF).



## Struktur File Submit

Submit dalam bentuk **Google Drive atau GitHub Repo** dan share link-nya melalui form yang akan disediakan.




## Dokumentasi Wajib Memuat
1. Judul Project
2. Deskripsi Singkat (latar belakang masalah & tujuan)
3. Fitur-Fitur Utama
4. Cara Menjalankan di Lokal
5. Preview Aplikasi (Screenshot / Demo)
6. Link Deploy Aplikasi
